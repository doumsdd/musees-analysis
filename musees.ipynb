```python
import numpy as np
import pandas as pd

df = pd.read_csv('liste-des-musees-de-france-2021_-_liste-des-musees-de-france-2021.csv')

df = df.tail(-1)
df.head(30)
```

```python
# Compte le nombre de musées distincts dans la colonne "name"
nombre_de_musees = df['name'].nunique()

print(nombre_de_musees)
```

```python
nombre_total_occurrences = df['name'].count()

print(nombre_total_occurrences)
```

```python
# Supprime les doublons basés sur la colonne "name" pour ne garder qu'un seul enregistrement par musée
df_unique_musees = df.drop_duplicates(subset='name', keep='first')

# Compte le nombre de musées uniques par ville dans la colonne "city"
comptage_musees_uniques_par_ville = df_unique_musees['city'].value_counts()

print(comptage_musees_uniques_par_ville)
```

```python
# Calculer le nombre moyen de musées par ville
nombre_moyen_musees_par_ville = comptage_musees_par_ville.mean()

print(nombre_moyen_musees_par_ville)
```

```python
# Calculer le nombre médian de musées par ville
nombre_median_musees_par_ville = comptage_musees_par_ville.median()

print(nombre_median_musees_par_ville)
```

```python
# Convertissez la colonne "tags" en une liste de mots descriptifs pour chaque musée
tag = df['tags'].apply(lambda x: x.split(':') if isinstance(x, str) else [])

# Excluez les valeurs de type float (NaN) et les valeurs vides lors de l'extraction des mots descriptifs
caracteres_a_supprimer = [';', 'type']
types_musees = [mot.strip() for mots in tag for mot in mots if mot.strip() and mot.strip().lower() not in caracteres_a_supprimer]

# Comptez le nombre de musées pour chaque type
comptage_musees_par_type = pd.Series(types_musees).value_counts()

# Calculez le nombre total de musées
nombre_total_musees = len(df)

# Calculez le pourcentage de chaque type de musée par rapport au nombre total de musées
pourcentage_musees_par_type = (comptage_musees_par_type / nombre_total_musees) * 100

print(pourcentage_musees_par_type)
```

```python
masque_chateau = df['name'].str.startswith('Château')

# Comptez le nombre de musées dont le nom commence par "Château"
nombre_musees_chateau = masque_chateau.sum()

print(nombre_musees_chateau)
```

```python
# booléen pour filtrer les valeurs non nulles (non NaN) dans la colonne "website"
website_non_null = df['website'].notna()

# filtrer les musées avec une adresse de site web valide
musees_avec_website = df[website_non_null]

# Comptez le nombre de musées ayant une adresse de site web valide
nombre_musees_avec_website = len(musees_avec_website)

print(nombre_musees_avec_website)
```

```python
# Convertir la colonne "postal code" en chaînes de caractères
df['postal_code'] = df['postal_code'].astype(str)

# Supprimer les espaces et autres caractères non numériques
df['postal_code'] = df['postal_code'].str.replace(r'\D', '', regex=True)

# Supprimer les codes postaux avec plus de 5 chiffres
df = df[df['postal_code'].str.len() <= 5]

# Extraire les deux premiers chiffres du code postal pour obtenir le département
df['departement'] = df['postal_code'].str[:2].astype(int)

# Utiliser groupby pour regrouper les musées par département
groupe_musees_par_departement = df.groupby('departement')

# Compter le nombre de musées dans chaque département et trouver le département avec le nombre le plus élevé de musées
departement_plus_de_musees = groupe_musees_par_departement.size().idxmax()

print("Le département français avec le plus de musées est le", departement_plus_de_musees)
```

```python
# Compter le nombre de musées dans chaque département et trouver le département avec le nombre le plus bas de musées
departement_moins_de_musees = groupe_musees_par_departement.size().idxmin()

print("Le département français avec le moins de musées le", departement_moins_de_musees)
```

```python
# masque booléen indiquant si "Napoléon" est présent dans le nom
masque_napoleon = df['name'].str.contains('Napoléon', case=False)

# Compter le nombre de musées ayant "Napoléon" dans leur nom
nombre_musees_napoleon = masque_napoleon.sum()

print("Le nombre de musées ayant 'Napoléon' dans leur nom est :", nombre_musees_napoleon)
```

```python
# Convertir la colonne "postal code" en chaînes de caractères
df['postal_code'] = df['postal_code'].astype(str)

# Supprimer les espaces et autres caractères non numériques
df['postal_code'] = df['postal_code'].str.replace(r'\D', '', regex=True)

# Extraire les deux premiers chiffres du code postal pour obtenir le département
df['departement'] = df['postal_code'].str[:2].astype(int)

# Utiliser groupby pour regrouper les musées par département
groupe_musees_par_departement = df.groupby('departement')

# Calculer le nombre total de musées
nombre_total_musees = len(df)

# Calculer le pourcentage de musées dans chaque code postal
pourcentage_musees_par_code_postal = (groupe_musees_par_departement / nombre_total_musees) * 100

print(pourcentage_musees_par_code_postal)
```

```python
# Utiliser loc pour trouver le musée du Louvre (en supposant que "Louvre" est dans la colonne "name")
louvre_data = df.loc[df['name'].str.contains('Louvre', case=False)]

# Si le DataFrame contient plusieurs musées avec "Louvre" dans le nom, vous pouvez en extraire le premier
if not louvre_data.empty:
    louvre_website = louvre_data.iloc[0]['website']
    print("Le website du musée du Louvre est :", louvre_website)
else:
    print("Le musée du Louvre n'a pas été trouvé dans le DataFrame.")
```
